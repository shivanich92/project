```python
import pandas as pd
import math
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Read the CSV file into a DataFrame 'df'
df = pd.read_csv('/content/Household energy bill data.csv')

# Check for missing values in the DataFrame 'df' and fill them with 0
df = df.fillna(0)

# Create a Series 'y' for the target variable 'amount_paid'
y = df['amount_paid']

# Create a DataFrame 'X' for the features
X = df[['num_rooms', 'num_people', 'housearea', 'is_ac', 'is_tv', 'is_flat',
        'ave_monthly_income', 'num_children', 'is_urban']]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2529)

# Create a Linear Regression model 'lr'
lr = LinearRegression()

# Fit the model on the training data
lr.fit(X_train, y_train)


# Predict 'y' values on the test data
y_pred = lr.predict(X_test)

# Create a scatter plot of actual vs predicted values
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Trades")
plt.ylabel("Predicted Trades")
plt.title("Actual Trades vs Predicted Trades")
plt.show()


# Linear Regression
model = LinearRegression()
model.fit(X, y)
r_sq = model.score(X, y)

var = ((y_test - y_pred)*(y_test - y_pred)).sum()
mse = (var/len(y_test))
print("MSE for Linear Regression: ",mse)

var = abs((y_test - y_pred)).sum()
mae = var/len(y_test)
print("Mean Absolute Error :",mae)

var = abs((y_test - y_pred)/y_test).sum()
mape = var/len(y_test) *100
print("Mean Absolute Error Percentage Error :",mape)

var = ((y_test - y_pred)*(y_test - y_pred)).sum()
rmse = math.sqrt(var/len(y_test))
print("RMSE for Linear Regression: ",rmse)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Read the CSV file from a ZIP archive into a DataFrame 'df'
df = pd.read_csv('/content/Household energy bill data.csv')

# Check for missing values in the DataFrame 'df' and fill them with 0
df = df.fillna(0)

# Create a Series 'y' for the target variable 'amount_paid'
y = df['amount_paid']

# Create a DataFrame 'X' for the features
X = df[['num_rooms', 'num_people', 'housearea', 'is_ac', 'is_tv', 'is_flat',
        'ave_monthly_income', 'num_children', 'is_urban']]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2529)

# Ridge Regression
rr = Ridge(alpha=0.01)
rr.fit(X_train, y_train)
y_pred_rr = rr.predict(X_train)

# Calculate RMSE and R2 score for Ridge Regression
rmse_rr = np.sqrt(mean_squared_error(y_train, y_pred_rr))
r2_rr = r2_score(y_train, y_pred_rr)
mae_rr = mean_absolute_error(y_train, y_pred_rr)
print("Mean Absolute Error for Ridge Regression: ",mae_rr)

mape_rr = np.mean(np.abs((y_train - y_pred_rr)/y_train)*100)
print("Mean Absolute percentage Error: ",mape_rr)
print(f"RMSE for Ridge Regression: {rmse_rr}")
print(f"R2 Score for Ridge Regression: {r2_rr}")
```
